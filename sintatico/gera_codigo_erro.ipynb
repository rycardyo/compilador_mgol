```python
import pandas as pd 
import numpy as np
```

```python
csv = pd.read_csv('./Shift-Reduce.csv')
csv.head()
```

```python
csv.fillna('E-00',inplace=True)
csv.head()
```

```python
for c in csv.columns:
    csv[c] = csv.apply(lambda x: x[c].replace('E-00','E-{}'.format(x.name)), axis = 1)
csv
```

```python
csv.to_csv('./Shift-Reduce.csv',index=False, index_label=False)
```
